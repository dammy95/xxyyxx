### Time and Energy!

The `date-time` column records the date and exact time of each record. This allows us to examine how energy is used as it relates with the months, weeks, days, etc.
To do this, I add the following columns to the dataframe: `month`, `hour`, `weekday`. With these new columns, we can examine the energy usage against the various months, hour of day, or weekday.

```python
months_of_year = ['January', 'February', 'March', 'April', 'May']
days_of_week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

# Change date-time field to python datetime object.
df['date-time'] = pd.to_datetime(df['date-time'], format='%Y-%m-%d %H:%M:%S')

# Make sure we're able to sort according to month of the year
df['month'] = df['date-time'].dt.month_name()
df['month'] = pd.Categorical(df['month'], months_of_year) 

# Make sure we're able to sort according to day of week
df['week'] = df['date-time'].dt.weekday_name
df['week'] = pd.Categorical(df['week'], days_of_week) 

df['hour'] = df['date-time'].dt.hour

df.head()
```

Let's examine how energy is used per month, against our target variable (`appliance (Wh)`), amongst others:

```python
months = df['month'].unique()
appliance_data = df.groupby('month')['appliances (Wh)'].mean()
lights_data = df.groupby('month')['lights (Wh)'].mean()
visibility_data = df.groupby('month')['visibility'].mean()
outdoor_data_temp = df.groupby('month')['northside-outdoor temperature'].mean() 
outdoor_data_hum = df.groupby('month')['northside-outdoor humidity'].mean() 
visibility_data = df.groupby('month')['visibility'].mean() 
pressure_data = df.groupby('month')['pressure'].mean() 
windspeed_data = df.groupby('month')['windspeed'].mean() 
tdewpoint_data = df.groupby('month')['t-dew-point'].mean() 

concat_list = [
    appliance_data, lights_data, outdoor_data_temp, outdoor_data_hum, 
    visibility_data, pressure_data, windspeed_data, tdewpoint_data
]
monthly_data = pd.concat(concat_list, axis=1).reset_index()
data_list = [
    'appliances (Wh)', 'lights (Wh)', 'northside-outdoor temperature', 
    'northside-outdoor humidity', 'visibility', 'pressure', 'windspeed', 
    't-dew-point'
]

monthly_plot_1 = sns.PairGrid(
    monthly_data, 
    x_vars=data_list[:4],
    y_vars=['month'], 
    height=4
)
monthly_plot_1.map(sns.barplot)

monthly_plot_2 = sns.PairGrid(
    monthly_data, 
    x_vars=data_list[4:8],
    y_vars=['month'], 
    height=4
)
monthly_plot_2.map(sns.barplot)
```

The barplots above shows how eight variables changes with the five different months in which the recordings was taken. By grouping the data according to the months and taking the mean of the different variables, we see exactly how these variables relate with the months.

1. The `appliances (Wh)` variable (our target value), which shows the amount of energy used up by the appliances in the house, doesn't show any significant change throughtout the months. It shows roughly the same amount of energy used throughout the months of January, up to May.

2. The `lights (Wh)` variable shows the amount of energy used up by the lights in the home, and shows a general decrease as the seasons change from winter (January, February) to spring (March, April, May).

3. The `northside-outdoor temperature` and `t-dew-point` shows a general increase for the same reasons as the `lights (Wh)` variable: the seasonal change from winter to spring. `t-dew-point` is affected by temperature, which increases as temperature increases (and vice versa).

4. Seasonal change is also attributed to the decrease in humidity for the `northside-outdoor humidity`.

5. `visibility` and `pressure` shows no significant increase or decrease throughout the months. 

6. `windspeed` decreases from January to May.

We can also investigate how energy is used during the different hours of the day, as well as the days of the week. To do this, I plot a heatmap showing how the hours of the day and the days of the week are related to the `appliances (Wh)` variable:

```python
df['appliances (Wh)'].mean()
```

```python
def plot_heat_map_energy(column, title):
    """
    Helper function to create heatmap of energy use during day and hours.
    """
    months_of_year = ['January', 'February', 'March', 'April', 'May']

    fig, axs = plt.subplots(2, 3, figsize=(20,20))
    fig.suptitle(title, fontsize=18)
    for ax, month in zip(axs.flatten(), months_of_year):
        weekly_data = df[df['month'] == month][['week', 'hour', column]]
        weekly_data = pd.pivot_table(weekly_data, values=column, index=['hour'], columns=['week'])
        ax.set(title=month)
        sns.heatmap(weekly_data, square=True, ax=ax, cmap='OrRd', linewidth=1)
    axs.flat[-1].set_visible(False)
```

```python
plot_heat_map_energy('appliances (Wh)', 'Fig 4: Heatmap of Appliance usage (Wh)')
```

The heatmap in `Fig 4` above shows the use of appliances (Wh) during the days of the weeks and hours of the day for the five months the data was recorded.

As shown (AUA  Appliance Usage Activity):

1. For all months, a higher proportion of appliance usuage activity (AUA) occurs on Sunday and Saturday, the two weekend days.
2. For all months, the lowest AUA occurs between the hours of `9pm` to `7am` (i.e sleeping hours).
3. February and March show a high AUA on Mondays between the hours of `10am` to `6pm`.
4. For January, April, and May, the weekday (i.e Monday  Friday) hours of `7am` to `9pm` have a lower AUA than those of weekends.
5. On weekdays, the hours of `5  6 pm` have the highest AUA.

There is certainly a relationship between AUA, month, and hour of the day. I believe there is sufficient evidence from the heatmap to consider these three features into a machine learning model to improve performance.

```python
plot_heat_map_energy('lights (Wh)', 'Fig 5: Heatmap of Lights usage (Wh)')
```

The heatmap in `Fig 5` above shows the use of Lights (Wh) usage activity (LUA) during the days of the weeks and hours of the day for the five months the data was recorded.

As shown:

1. January and Febuary have a higher LUA than other months (see barplot on LUA against months). This could be due to the seasons: Januar and February (winter) is considerably darker than March, April, and May, when spring starts, and the days are considerably longer.
2. For the months of January, and February, the LUA is considerably higher during night times (5pm  12am). This is expected as night is darker than day. The only exception, for both months, to this trend is Sunday, which reports a high LUA from 8am to 11pm.
3. For all months, 12am to 7am (sleeping hours) report the lowest LUA.
4. For all months, Mondays seem to have a disproportionately high LUA compared to other weekdays, especially during the 'active' night hours (i.e 6  11pm).

The heatmaps shows also shows a correlation between hours of the day, days of the week, and month with LUA.
